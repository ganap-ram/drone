### Parameters ###
\begin{equation*}
\begin{aligned}

{u_0}, {v_0} &= Starting\hspace{1pt}position\hspace{1pt}of\hspace{1pt}the\hspace{1pt}drone\\
u_{J+1}, v_{J+1} &= Ending\hspace{1pt}position\hspace{1pt}of\hspace{1pt}the\hspace{1pt}drone\\
\end{aligned}    
\end{equation*}

### Variables/Simplifications ###
\begin{equation*}
\begin{aligned}
{u_j}, {v_j} &= j_{th}\hspace{1pt}position\hspace{1pt}of\hspace{1pt}drone\hspace{4pt}j = {1..J}\\
{a_j} &= {u_j - x_j}\\
{b_j} &= {v_j - y_j}\\

{m_j} &= u_{j} - u_{j+1}\\
{n_j} &= v_{j} - v_{j+1}\\
M_j &= {m_j}^2 + {n_j}^2 \\
H_j &= {M_j}^{3/2} \\
q &= \frac{p}{2} - 1\\
r &= 1/2\\
\end{aligned}    
\end{equation*}

### Gradient Equations ###
\begin{equation}\label{eq:def_2_pds}
\begin{aligned}   
\frac{\partial f}{\partial u_j} &= p{a_j}({a_j}^2 + {b_j}^2)^q \hspace{8pt} (where \hspace{4pt} j=1....J)\\
\frac{\partial f}{\partial v_j} &= p{b_j}({a_j}^2 + {b_j}^2)^q \hspace{8pt} (where \hspace{4pt} j=1....J)\\
\frac{\partial g}{\partial u_j} &= -2rm_{j-1}(m_{j-1}^{2} + n_{j-1}^{2})^{r-1} + 2rm_{j}(m_{j}^{2} + n_{j}^{2})^{r-1} \hspace{8pt} (where \hspace{4pt} j=1....J)\\
\frac{\partial g}{\partial v_j} &= -2rn_{j-1}(m_{j-1}^{2} + n_{j-1}^{2})^{r-1} + 2rn_{j}(m_{j}^{2} + n_{j}^{2})^{r-1} \hspace{8pt} (where \hspace{4pt} j=1....J)\\
\frac{dg}{ds} &= -1
\end{aligned}    
\end{equation}\\

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import warnings
import random as rand
import pandas as pd
from scipy.spatial.distance import cdist
from python_tsp.exact import solve_tsp_dynamic_programming
from timeit import default_timer as timer

warnings.simplefilter('default')

def timedTsp(dists):
    def tsp_wrapper():
        return solve_tsp_dynamic_programming(dists)
    
    # Time the execution of the wrapper function
    execution_time = timeit.timeit(tsp_wrapper, number=1)
    
    # Call the wrapper once to get the return value
    perm, dists = tsp_wrapper()
    
    # Return both the execution time and the return value
    return perm, dists, execution_time

def arrange_points(XS, YS):
    points = np.vstack((XS, YS))
    points = points.T

    dists = cdist(points, points)

    time = 0
    perm, dists = solve_tsp_dynamic_programming(dists)
    # perm, dists, time = timedTsp(dists)

    XS_Solved = XS[perm]
    YS_Solved = YS[perm]
    
    return XS_Solved, YS_Solved, time

def gf_values(uj, vj, u0, v0, xj, yj, P):
    J = len(uj)

    ## Finding f
    aj = uj - xj
    bj = vj - yj
    f = np.sum((aj**2 + bj**2)**(P/2))
    
    ## Finding g
    uj_p_1 = np.array(uj)
    vj_p_1 = np.array(vj)   
    uj_p_1[0] = 0
    vj_p_1[0] = 0    
    uj_p_1 = np.roll(uj_p_1, -1)
    vj_p_1 = np.roll(vj_p_1, -1)

    # Build our mj and nj    
    mj_tmp = np.zeros(J + 1)
    nj_tmp = np.zeros(J + 1)
    mj_tmp[0] = u0 - uj[0]
    nj_tmp[0] = v0 - vj[0]
    mj_tmp[1:J+1] = uj - uj_p_1
    nj_tmp[1:J+1] = vj - vj_p_1
    
    g = np.sum((mj_tmp**2 + nj_tmp**2)**0.5)
    
    return (g,f)

def find_initial_values(xj, yj, u0, v0, P):
    Q = (P / 2) - 1
    R = 1/2
    e = 0.1
    J = len(xj)
    xj_p_1 = np.array(xj)
    yj_p_1 = np.array(yj)
    xj_p_1[0] = 0
    yj_p_1[0] = 0
    xj_p_1 = np.roll(xj_p_1, -1)
    yj_p_1 = np.roll(yj_p_1, -1)

    # Build our mj and nj
    mj_tmp = np.zeros(J + 1)
    nj_tmp = np.zeros(J + 1)
    mj_tmp[0] = u0 - xj[0]
    nj_tmp[0] = v0 - yj[0]
    mj_tmp[1:J+1] = xj - xj_p_1
    nj_tmp[1:J+1] = yj - yj_p_1

    mj = mj_tmp[1:J+1]
    nj = nj_tmp[1:J+1]

    mj_m_1 = mj_tmp[0:J]
    nj_m_1 = nj_tmp[0:J]

    mj_sq = mj**2
    nj_sq = nj**2
    mj_m_1_sq = mj_m_1**2
    nj_m_1_sq = nj_m_1**2

    Mj = mj_sq + nj_sq
    Mj_m_1 = mj_m_1_sq + nj_m_1_sq

    Mj_r_1 = Mj**(R-1)
    Mj_m_1_r_1 = Mj_m_1**(R-1)

    gu = -2*R*mj_m_1*Mj_m_1_r_1 + 2*R*mj*Mj_r_1
    gv = -2*R*nj_m_1*Mj_m_1_r_1 + 2*R*nj*Mj_r_1
    
    normg = np.sqrt(gu**2 + gv**2)
    g_factor = normg**(1/(P-1)-1)

    uj = xj - e * gu*g_factor
    vj = yj - e * gv*g_factor

    aj = uj - xj
    bj = vj - yj
    aj_sq = aj**2
    bj_sq = bj**2
    Aj = aj_sq + bj_sq
    Aj_q = Aj**Q

    fu = P*aj*Aj_q
    fv = P*bj*Aj_q

    Lambda = (np.sum(fu/gu) + np.sum(fv/gv))/(2*len(gu))

    return ((uj), (vj), Lambda)

def dudvdl_ds(SValue, uvl, P, u0, v0, xj, yj):
    Q = (P / 2) - 1
    R = 1/2    
    uv = uvl[0:len(uvl)-1]
    Lambda = uvl[len(uvl)-1]
    uv = np.array_split(uv, 2)
    J = len(uv[0])
    uj = uv[0]
    vj = uv[1]
    
    aj = uj - xj
    bj = vj - yj
    aj_sq = aj**2
    bj_sq = bj**2
    Aj = aj_sq + bj_sq
    Aj_q = Aj**Q
    Aj_q_m_1 = Aj**(Q - 1)
    
    # uj_plus_1 : the last doesn't have a next, same with vj_plus_1
    uj_p_1 = np.array(uj)
    vj_p_1 = np.array(vj)   
    uj_p_1[0] = 0
    vj_p_1[0] = 0    
    uj_p_1 = np.roll(uj_p_1, -1)
    vj_p_1 = np.roll(vj_p_1, -1)

    # Build our mj and nj    
    mj_tmp = np.zeros(J + 1)
    nj_tmp = np.zeros(J + 1)
    mj_tmp[0] = u0 - uj[0]
    nj_tmp[0] = v0 - vj[0]
    
    mj_tmp[1:J+1] = uj - uj_p_1
    nj_tmp[1:J+1] = vj - vj_p_1
    
    mj = mj_tmp[1:J+1]
    nj = nj_tmp[1:J+1]
    
    mj_m_1 = mj_tmp[0:J]
    nj_m_1 = nj_tmp[0:J]
    
    mj_sq = mj**2
    nj_sq = nj**2
    mj_m_1_sq = mj_m_1**2
    nj_m_1_sq = nj_m_1**2
    
    Mj = mj_sq + nj_sq
    Mj_m_1 = mj_m_1_sq + nj_m_1_sq
    
    Mj_r = Mj**(0.5)
    Mj_m_1_r = Mj_m_1**(0.5)
    
    Hj = Mj**(3/2)
    Hj_m_1 = Mj_m_1**(3/2)
    HH = Hj * Hj_m_1
    
    ## Create the T and S matrices that we can build independantly and interleave later
    S = np.zeros((J, 2*J))
    T = np.zeros((J, 2*J))
    C1 = np.zeros((J, 1))
    C2 = np.zeros((J, 1))  
    
    S[:,0] = Lambda*((Mj_m_1 - mj_m_1_sq)/Hj_m_1)
    S[:,1] = -Lambda*((mj_m_1*nj_m_1)/Hj_m_1)    
    S[:,2] = (2*P*Q*aj_sq*Aj_q_m_1 + P*Aj_q - (Lambda*((Hj*(Mj_m_1 - mj_m_1_sq) + Hj_m_1*(Mj - mj_sq))/HH)))    
    S[:,3] = (2*P*Q*aj*bj*Aj_q_m_1 + (Lambda*((mj_m_1*nj_m_1*Hj + mj*nj*Hj_m_1)/HH)))    
    S[:,4] = Lambda*((Mj - mj_sq)/Hj)   
    S[:,5] = -Lambda*((mj*nj)/Hj)
    
    T[:,0] = S[:,1]    
    T[:,1] = Lambda*((Mj_m_1 - nj_m_1_sq)/Hj_m_1)    
    T[:,2] = S[:,3]    
    T[:,3] = (2*P*Q*bj_sq*Aj_q_m_1 + P*Aj_q - (Lambda*((Hj*(Mj_m_1 - nj_m_1_sq) + Hj_m_1*(Mj - nj_sq))/HH)))    
    T[:,4] = S[:,5]    
    T[:,5] = Lambda*((Mj - nj_sq)/Hj)
    
    # Create C1 and C2
    W = (-mj_m_1*Mj_r + mj*Mj_m_1_r)/(Mj_r * Mj_m_1_r)
    C1[:,0] = W
    
    Z = (-nj_m_1*Mj_r +nj*Mj_m_1_r)/(Mj_r * Mj_m_1_r)
    C2[:,0] = Z

    # Adjust the co-efficients for the first and last points/cluster heads
    S[0,0] = S[0,1] = T[0,0] = T[0,1] = 0
    S[J-1,5] = S[J-1,4] = T[J-1,5] = T[J-1,4] = 0        
    
    # Roll the entries to form the co-effient matrix
    # This does loop, may be there is a way of vectorizing it!
    roll = -2
    for j in range(0, J):
        S[j,:] = np.roll(S[j,:], roll)
        T[j,:] = np.roll(T[j,:], roll)
        roll = roll + 2
        
    K = np.zeros((2*J + 1, 2*J + 1))
    C = np.zeros((2*J + 1, 1))
    for j in range(0, J):
        # Fill 2*J columns of each pairs of the rows
        K[2*j,0:2*J] = S[j,:]
        K[2*j + 1,0:2*J] = T[j,:]
        
        # Fill the last column of each pairs of the rows
        K[2*j,2*J] = -C1[j,0]
        K[2*j + 1,2*J] = -C2[j,0]
        
        # Fill the last row one by one
        K[2*J,2*j] = W[j]
        K[2*J,2*j + 1] = Z[j]
        
    C[2*J, 0] = -1
    
    dudvdls = np.linalg.solve(K, C)
    dus = np.array(dudvdls[range(0, 2*J, 2)])
    dvs = np.array(dudvdls[range(1, 2*J, 2)])
    dls = np.array([dudvdls[2*J]])
    
    dudvdls = np.reshape(np.concatenate((dus, dvs, dls)), 2*J + 1)
   
    return dudvdls

def rk4(N, H, uvl, P, u0, v0, xj, yj):
    # (*) first row in the result will be the initial points
    # (*) every iteration rows in the result has the staring and
    #     ending points appended, so that a circular path is formed.
    ret = np.zeros((N + 1, len(uvl) + 4))
    S = uvl[len(uvl)-1]
    J = len(xj)

    
    ret[0,0] = u0 # starts at arbitrary point
    ret[0,1:J+1] = uvl[0:J] 
    ret[0,J+1] = 0 # ends at 0
    ret[0,J+2] = v0 # starts at arbitrary point
    ret[0,J+2+1:J+2+1+J] = uvl[J:2*J]
    ret[0,J+2+1+J] = 0 # ends at 0
    ret[0,J+2+1+J+1] = uvl[2*J]
    
    for i in range(1, N+1):
        k1 = H * (dudvdl_ds(S, uvl, P, u0, v0, xj, yj))
        k2 = H * (dudvdl_ds(S+H/2, uvl+k1/2, P, u0, v0, xj, yj))
        k3 = H * (dudvdl_ds(S+H/2, uvl+k2/2, P, u0, v0, xj, yj))
        k4 = H * (dudvdl_ds(S+H, uvl+k3, P, u0, v0, xj, yj))
        k = (k1+2*k2+2*k3+k4)/6
        tmp = uvl + k
        uvl = tmp
        S = S + H
        # ret[i] = uvl
        ret[i,0] = u0 # starts at arbitrary point
        ret[i,1:J+1] = uvl[0:J] 
        ret[i,J+1] = 0 # ends at 0
        ret[i,J+2] = v0 # starts at arbitrary point
        ret[i,J+2+1:J+2+1+J] = uvl[J:2*J]
        ret[i,J+2+1+J] = 0 # ends at 0
        ret[i,J+2+1+J+1] = uvl[2*J]

    return ret

# The result uvl has the initial points included at the begining and end
# so that callers can see the points as a circular path
def Solve(P, xj, yj, u0, v0, H, NFactor):
    start_full = timer()
    
    # convert the input list to array
    xj = np.array(xj)
    yj = np.array(yj)
    
    # arnage the points to form the shortest path
    start = timer()
    xj, yj, t = arrange_points(xj, yj)
    elapsed_tsp = timer() - start
    
    # Find initial uj, vj and Lambda
    uvl = find_initial_values(xj, yj, u0, v0, P)
    L = np.array([uvl[2]])
    
    # Find the g, f values and the range
    (g, f) = gf_values(uvl[0], uvl[1], u0, v0, xj, yj, P)
    N = (int)(NFactor * g//H)

    # Create our variable list as an array
    uvl = np.concatenate((uvl[0], uvl[1], L))
    
    # Invoke RK method to approximate our variables
    start = timer()    
    uvl = rk4(N, H, uvl, P, u0, v0, xj, yj)
    elapsed_rk4 = timer() - start

    elapsed_full = timer() - start_full
    
    return uvl, (elapsed_tsp, elapsed_rk4, elapsed_full)

def PlotResults(inputs, results, xj_yj_sizes, tsp_times, rk4_times, full_times, plot_percent=7):
    ri = 0
    rows = math.ceil(len(results) / 2)
    cols = 2
    
    fig_base_rows = 14
    fig_base_cols = 24
    
    fig1, axes_paths = plt.subplots(rows, cols, figsize=(fig_base_rows, fig_base_cols))
    fig2, axes_gf = plt.subplots(1, 2, figsize=(fig_base_rows, fig_base_cols//5))
    fig3, axes_g_vs_f = plt.subplots(1, 1, figsize=(fig_base_rows, fig_base_cols//5))
    fig4, axis_path_len_diff = plt.subplots(1, 1, figsize=(fig_base_rows, fig_base_cols//5))
    fig5, axis_energy_diff = plt.subplots(1, 1, figsize=(fig_base_rows, fig_base_cols//5))
    fig5, ax_times = plt.subplots(1, 1, figsize=(fig_base_rows, fig_base_cols//5))
    
        
    axis_path_len_diff.set_title("iterations vs path-length difference")
    axis_path_len_diff.set_xlabel("iterations")
    axis_path_len_diff.set_ylabel("path length difference")        

    axis_energy_diff.set_title("iterations vs energy difference")
    axis_energy_diff.set_xlabel("iterations")
    axis_energy_diff.set_ylabel("energy difference")        
    
    axes_gf[0].set_title("iterations vs tour_length - path_length")
    axes_gf[0].set_xlabel("iterations")
    axes_gf[0].set_ylabel("tour_length - path_length")        
        
    axes_gf[1].set_title("iterations vs total-energy")
    axes_gf[1].set_xlabel("iterations")
    axes_gf[1].set_ylabel("total energy")
    
    axes_g_vs_f.set_title("tour_length - path_length vs (total-energy)^(1/P)")
    axes_g_vs_f.set_xlabel("tour_length - path_length")
    axes_g_vs_f.set_ylabel("(total-energy)^(1/P)")
            
    for result in results:
        # Grab the lambdas
        ls = result[:,len(result[0])-1]
        
        # Our inputs - cluster heads and starting points, cluster head needs
        # to be arranged, as we need to find the g and f for plotting
        (xj, yj, t) = arrange_points(np.array((inputs[ri])[0]), np.array((inputs[ri])[1]))
        
        # Our arbitrary starting point
        (u0, v0) = (inputs[ri][2])[0], (inputs[ri][2])[1]
        
        # P for the power function
        P = (inputs[ri][3])
        
        # Grab the uj,vjs
        ujvjs = result[:,0:len(result[0])-1]        
        J = len(ujvjs[0])//2
        
        # Reshape the array so that we have list ujs and list of vjs for all
        # iteration results
        ujvjs = np.reshape(ujvjs, (len(ujvjs), 2, J))
        
        # All the uj vjs that rk4 apprxomated
        ujs = ujvjs[:,0,1:J-1]
        vjs = ujvjs[:,1,1:J-1]

        # This is the initial uj, vjs
        init_ujs = ujvjs[0,0,1:J-1]
        init_vjs = ujvjs[0,1,1:J-1]
        
        # This is the final uj, vjs
        final_ujs = ujvjs[len(ujvjs)-1,0,1:J-1]
        final_vjs = ujvjs[len(ujvjs)-1,1,1:J-1]
        
        # For showing the path the drone takes, we are not plotting all the points,
        # as the plot can look cluttered but equally spaced 7% of the solution points
        plot_count = (len(result) * plot_percent) // 100
        plot_count = np.minimum(plot_count, 100)
        indices = np.linspace(0, len(ls), plot_count, dtype=int)
        indices[len(indices) - 1] = indices[len(indices) - 1] - 1
        
        # This the list of iterations we would like to plot the entire paths
        plot_list = ujvjs[indices]

        # print("final ujs = ", np.round(final_ujs, decimals=2))
        # print("final_vjs = ", np.round(final_vjs, decimals=2))
        
        # Plot the drone path optimization
        for i in range(len(plot_list)):
            axes_paths[ri//2,ri%2].plot(plot_list[i,0], plot_list[i,1], alpha=0.7)
            
        # Plot Cluster heads, Initial points, the final points
        # the plotting order matters!
        axes_paths[ri//2,ri%2].scatter(init_ujs, init_vjs, label='initial uj,vj', marker='o', s=100, color='orange', alpha=1)
        axes_paths[ri//2,ri%2].scatter(xj, yj, label='cluster heads xj,yj', marker='x', s=100)        
        axes_paths[ri//2,ri%2].scatter(0, 0, label='ending point', marker='o', color='red', s=200)
        axes_paths[ri//2,ri%2].scatter(u0, v0, label='starting point', marker='D', color='#008800', s=80)
        axes_paths[ri//2,ri%2].scatter(final_ujs, final_vjs, label='final uj,vj', marker='*', color='m', s=80)
        
        # axes_paths[ri//2,ri%2].legend(fontsize='small')
        axes_paths[ri//2,ri%2].set_xlabel("uj")
        axes_paths[ri//2,ri%2].set_ylabel("vj")
        
        axes_paths[ri//2,ri%2].set_title("Drone paths iterations: " + "test-" + str(ri))
        axes_paths[ri//2,ri%2].grid(True)        
        
        #print(len(results[ri]))
        gs = np.zeros(len(results[ri]))
        fs = np.zeros(len(results[ri]))
        
        g_diff = np.zeros(len(results[ri]) - 1)
        f_diff = np.zeros(len(results[ri]) - 1)
        
        for gi in range(len(results[ri])):
            gfs = gf_values(ujs[gi], vjs[gi], u0, v0, xj, yj, P)
            fs[gi] = gfs[1]            
            if gi == 0:
                g0 = gfs[0]
            else:
                gs[gi] = g0 - gfs[0]
            
        g_diff = np.absolute(np.ediff1d(gs))
        f_diff = np.absolute(np.ediff1d(fs))

        axes_gf[0].plot(gs, label="test-"+str(ri))
        axes_gf[1].plot(fs, label="test-"+str(ri))        
        axes_g_vs_f.plot(gs, (fs)**(1/P), label="test-"+str(ri))
        axis_path_len_diff.plot(g_diff, label="test-"+str(ri)) 
        axis_energy_diff.plot(f_diff, label="test-"+str(ri)) 
        
        ri = ri + 1

    # fig1.legend(fontsize='small')        
    # lines_labels = [ax.get_legend_handles_labels() for ax in fig1.axes]
    # lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
    # fig1.legend(lines, labels)
    lines_labels = [fig1.axes[0].get_legend_handles_labels()]
    lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
    # fig1.legend(lines, labels, loc='upper center', ncols=5)
    fig1.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.25, 0.42, 0.5, 0.5), ncols=5)

    axis_path_len_diff.legend()    
    axis_energy_diff.legend()
    axes_gf[0].legend()
    axes_gf[1].legend()
    axes_g_vs_f.legend()            
    
    axis_path_len_diff.grid(True)
    axis_energy_diff.grid(True)
    axes_gf[0].grid(True)
    axes_gf[1].grid(True)
    axes_g_vs_f.grid(True)           
    
    ax_times.plot(xj_yj_sizes, tsp_times, color='r', label='points arrangement runtime')
    ax_times.plot(xj_yj_sizes, rk4_times, color='g', label='rungekutta solver runtime')
    ax_times.plot(xj_yj_sizes, full_times, color='b', label='overall runtime')
    ax_times.legend()
    ax_times.grid(True)
    ax_times.set_xlabel("number of cluster heads")
    ax_times.set_ylabel("time taken (seconds)")
    ax_times.set_title("number of cluster heads vs runtime")
    
    plt.show()
```
